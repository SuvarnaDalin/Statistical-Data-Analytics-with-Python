Prob: Goodness of Fit Test: Parking Garage

```python
data = pd.read_excel('G:/IIT - nptel/Data Analytics with Python/data/P_distribution.xlsx')
data
```

```python
obs_freq = data['Frequency']
total_arrival = 600
total_time_period = 100
mu = total_arrival/ total_time_period
```

```python
exp_freq = []
for i in range(len(obs_freq)):
    E_Freq = 100 * poisson.pmf(i, mu)
    exp_freq.append(E_Freq)
```

```python
exp_freq
```

```python
exp_freq_round_off = [round(elem, 2) for elem in exp_freq]
exp_freq_round_off
```

```python
df = pd.DataFrame(list(zip(obs_freq, exp_freq_round_off)), columns = ['obs_freq', 'exp_freq'])
df
```
