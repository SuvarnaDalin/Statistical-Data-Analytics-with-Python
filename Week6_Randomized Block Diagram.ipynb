```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

Prob1:vWork Station Design Comparison

```python
df = pd.read_excel('G:/IIT - nptel/Data Analytics with Python/data/RBD.xlsx')
```

```python
df
```

```python
data = pd.melt(df.reset_index(), id_vars = ['index'], value_vars = ['System A', 'System B', 'System C'])
data.columns = ['index', 'treatments', 'value']
data
```

```python
model = ols('value ~ C(treatments)', data = data).fit()
anova_table = sm.stats.anova_lm(model, typ = 1) # lm =mlinear model, typ 1 => 1 way anova, type2 => 2 way anova
anova_table
```

Considering Blocking: Solving RBD

```python
bdata = pd.melt(df.reset_index(), id_vars = ['index'], value_vars = ['System A', 'System B', 'System C'])
bdata.columns = ['blocks', 'treatments', 'value']
bdata
```

```python
bmodel = ols('value ~ C(blocks) + C(treatments)', data = bdata).fit()
banova_table = sm.stats.anova_lm(bmodel, typ = 1)
banova_table
```

Prob2: Effect of 4 chemicals on fabric Strength

```python
df1 = pd.read_excel('G:/IIT - nptel/Data Analytics with Python/data/rbd2.xlsx')
df1
```

```python
ch_data = pd.melt(df1.reset_index(), id_vars = ['index'], value_vars = ['chem1', 'chem2', 'chem3', 'chem4'])
ch_data.columns = ['index', 'treatments', 'value']
ch_data
```

```python
ch_model = ols('value ~ C(treatments)', data = ch_data).fit()
ch_anova = sm.stats.anova_lm(ch_model, typ = 1)
ch_anova
```

Prob2: With Blocking

```python
dch_data = pd.melt(df1.reset_index(), id_vars = ['index'], value_vars = ['chem1', 'chem2', 'chem3', 'chem4'])
dch_data.columns = ['Fabric_samples', 'Chemical_types', 'value']
dch_data
```

```python
dch_model = ols('value ~ C(Fabric_samples) + C(Chemical_types)', data = dch_data).fit()
dch_anova = sm.stats.anova_lm(dch_model, typ = 1)
dch_anova
```
